```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
% matplotlib inline
```

```python
dt = pd.read_csv(r'NASA\\temprature.csv')
```

```python
dt.head()
```

```python
dt = dt.set_index("Source")
dt = dt.drop("GCAG", axis=0)
```

```python
dt.head()
```

```python
dt = dt.iloc[::-1]
```

```python
x = 0
tempp_Grth = list()
for y in dt['Mean']:
    tempp_Grth.append(y-x)
    x = y
```

```python
tempp_Grth
```

```python
dt['Temprature Growth'] = tempp_Grth
```

```python
dt
```

```python
dt.to_csv(r'NASA\\tempratureNew.csv')
```

```python
dsea = pd.read_csv(r'NASA\\sea level.csv')
```

```python
dsea.head()
```

```python
dsea = dsea.drop('NOAA Adjusted Sea Level',1)
```

```python
dsea.head()
```

```python
x = 0
se_Grth = list()
for y in dsea['CSIRO Adjusted Sea Level']:
    if (x!=0):
        se_Grth.append((y-x)/x)
    else :
        se_Grth.append(0)
    x = y
```

```python
dsea['Sea Level Growth'] = se_Grth
```

```python
dsea
```

```python
dsea.rename(columns={"CSIRO Adjusted Sea Level": "Sea Level"})
```

```python
dsea.to_csv(r'NASA\\Sea Level New.csv')
```

```python
dpop = pd.read_csv(r'NASA\\population.csv')
```

```python
years = []
for x in range(1960, 2019):
    years.append(x)
```

```python
years
```

```python
data = {'Year': [], 
        'Population': []} 
dnew = pd.DataFrame(data)
```

```python
population = []
for index, row in dpop.iterrows():
    for x in range(1960, 2019):
        population.append(row['Y'+str(x)])
    data = {'Year': years, 
        'Population': population}
    dnew = pd.DataFrame(data)
    dnew.to_csv(r'NASA\\new\\'+str(row['Country Name'])+'.csv')
    dnew = dnew.iloc[0:0]
    population.clear() 
```
